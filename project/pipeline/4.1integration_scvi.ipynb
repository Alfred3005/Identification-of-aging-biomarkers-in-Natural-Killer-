```python
#11/10/2024 aqui se modifica para leer solamenre el dataset correspondiente a tipos celulares

import os
import scanpy as sc
import anndata
import numpy as np
import scvi
import logging
import scib




scvi.settings.dl_num_workers = 14

# Configurar el logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def read_and_combine_h5ad(directory):
    logging.info(f"Leyendo archivos h5ad desde el directorio: {directory}")
    adata= sc.read_h5ad(directory)
    logging.info(f"adata shape: {adata.shape}")
    
    return adata


# Leer y combinar archivos h5ad
directory = '/app/project/pipeline_colombia/3.5normalized&scaled_by_group/processed_filtered_nk cells_data.h5ad'
combined_adata = read_and_combine_h5ad(directory)


# Configurar anndata para scVI
logging.info("Configurando anndata para scVI")
scvi.model.SCVI.setup_anndata(
    combined_adata,
    layer="counts",
    batch_key='short_title',
    categorical_covariate_keys= ['assay', 'self_reported_ethnicity', 'sex']
)

# Entrenar modelo scVI
logging.info("Entrenando modelo scVI")
scvi_model = scvi.model.SCVI(combined_adata, n_latent=25, n_layers=2, n_hidden=128, gene_likelihood="nb")
scvi_model.train(max_epochs=600)

# Guardar el modelo scVI
scvi_model_save_path = '/app/project/pipeline_colombia/4.models/scvi_model'
scvi_model.save(scvi_model_save_path, overwrite=True)

# Obtener representaciones latentes y visualización
SCVI_LATENT_KEY = "X_scVI"
combined_adata.obsm[SCVI_LATENT_KEY] = scvi_model.get_latent_representation()

logging.info("Calculating neighbors")
sc.pp.neighbors(combined_adata, use_rep=SCVI_LATENT_KEY)

logging.info("Running Leiden algorithm")
sc.tl.leiden(combined_adata, flavor="igraph", n_iterations=2, directed=False)

# Visualización UMAP
logging.info("Calculating UMAP")
sc.tl.umap(combined_adata, min_dist=0.1)

# Visualización y guardado de gráficos
embedding_plot_path_sample = '/app/project/pipeline_colombia/4.models/scvi_model/scvi_embedding_sample.png'
embedding_plot_path_cell_type = '/app/project/pipeline_colombia/4.models/scvi_model/scvi_embedding_cell_type.png'

logging.info("Guardando gráfico de muestra")
sc.pl.umap(
    combined_adata,
    color=["short_title", "leiden"],
    frameon=False,
    ncols=1,
    save=True
)

logging.info("Guardando gráfico por tipo de célula")
sc.pl.umap(combined_adata, color=["cell_type"], frameon=False, ncols=1, save=True)

# Guardar resultados de scVI
output_path = '/app/project/pipeline_colombia/4.models/scvi_model/scvi_integrated.h5ad'
logging.info(f"Guardando resultados en: {output_path}")
combined_adata.write(output_path)

logging.info("Finalización del script")
  
```
