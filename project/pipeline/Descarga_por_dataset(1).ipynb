```python
import h5py
import numpy as np
import anndata

def map_age_stage(stage):
    infant_labels = [
        'infant stage', 'newborn human stage', 'adolescent stage',
        '2-5 year-old child stage', '6-12 year-old child stage'
    ]
    adult_labels = [
        'human adult stage', 'third decade human stage',
        'fourth decade human stage', 'fifth decade human stage'
    ]
    aged_labels = [
        'human aged stage', 'sixth decade human stage',
        'seventh decade human stage', '80 year-old and over human stage'
    ]

    if stage in infant_labels:
        return 'Infantiles'
    if stage in adult_labels:
        return 'Adultos'
    if stage in aged_labels:
        return 'Ancianos'

    if 'year-old human stage' in stage:
        age = int(stage.split('-')[0])
        if 0 <= age <= 21:
            return 'Infantiles'
        elif 21 < age <= 55:
            return 'Adultos'
        else:
            return 'Ancianos'

    return 'Otros'

# Leer el archivo .h5ad en un objeto Anndata
#adata = anndata.read_h5ad("path_to_your_file/filtrado_test0908(1).h5ad")

# Aplicar el mapeo a la columna 'development_stage'
adata.obs["age_group"] = adata.obs["development_stage"].map(map_age_stage)

# Guardar el objeto Anndata modificado
adata.write("mapeado_age_test_2708.h5ad")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Para obtener una paleta de colores pastel

# Obtener los datos
grupos_edad_cuenta = adata.obs['age_group'].value_counts()

# Función personalizada para mostrar porcentaje y cantidad
def func(pct, allvalues):
    absolute = int(pct / 100. * np.sum(allvalues))
    return f"{pct:.1f}%\n({absolute:d})"

# Obtener una paleta de colores pastel
colores = sns.color_palette("pastel", len(grupos_edad_cuenta))

# Configurar explode para resaltar 'Infantiles'
explode = [0.2 if label == 'Infantiles' else 0 for label in grupos_edad_cuenta.index]

# Crear la gráfica de pastel
plt.figure(figsize=(12, 7))
wedges, texts, autotexts = plt.pie(grupos_edad_cuenta,
                                   labels=grupos_edad_cuenta.index,
                                   autopct=lambda pct: func(pct, grupos_edad_cuenta),
                                   startangle=140,
                                   colors=colores,
                                   explode=explode)

# Mapeo de etiquetas a sus descripciones
label_map = {
    "Infantiles": "Infantiles 0-21",
    "Adultos": "Adultos 22-55",
    "Ancianos": "Ancianos 56 en adelante"
}

# Ordenar las etiquetas de la leyenda según su aparición en grupos_edad_cuenta
leyenda_labels = [label_map[label] for label in grupos_edad_cuenta.index]

# Añadir la leyenda en la parte inferior
plt.legend(wedges, leyenda_labels, loc='upper center', bbox_to_anchor=(0.5, -0.15))

plt.title('Distribución por Grupo de Edad')
plt.show()
```
