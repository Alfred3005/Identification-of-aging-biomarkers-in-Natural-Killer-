```python
import os
import pegasus as pg
import ddqc
import matplotlib.pyplot as plt
import numpy as np

def plot_qc_metrics(df_qc, output_path):
    """Genera y guarda las gráficas de las métricas de control de calidad."""
    fig, axs = plt.subplots(2, 2, figsize=(15, 10))
    
    # Crear gráficos de ejemplo, puedes personalizarlos según tus necesidades
    df_qc['n_counts'].hist(ax=axs[0, 0])
    axs[0, 0].set_title('Total Counts')
    
    df_qc['n_genes'].hist(ax=axs[0, 1])
    axs[0, 1].set_title('Number of Genes by Counts')
    
    df_qc['percent_mito'].hist(ax=axs[1, 0])
    axs[1, 0].set_title('Percentage of Mitochondrial Counts')
    
    df_qc['percent_ribo'].hist(ax=axs[1, 1])
    axs[1, 1].set_title('Percentage of Ribosomal Counts')
    
    plt.tight_layout()
    plt.savefig(output_path)
    plt.show()

def process_quality_control(input_directory, qc_output_directory, filtered_output_directory):
    # Asegurarse de que los directorios de salida existan
    os.makedirs(qc_output_directory, exist_ok=True)
    os.makedirs(filtered_output_directory, exist_ok=True)

    for filename in os.listdir(input_directory):
        if filename.endswith(".h5ad"):
            print(f"Processing {filename}...")
            filepath = os.path.join(input_directory, filename)
            data = pg.read_input(filepath, genome='hg19')
            
            # Check if raw counts are in .X
            if not np.issubdtype(data.X.dtype, np.integer):
                print(f"Warning: {filename} does not contain raw counts in .X")
            else:
                print(f"{filename} contains raw counts in .X")
            
            # Realizar el control de calidad y obtener el DataFrame de QC
            df_qc = ddqc.ddqc_metrics(data, return_df_qc=True, display_plots=True)
            
            # Guardar el DataFrame de QC
            qc_output_path = os.path.join(qc_output_directory, f"qc_metrics_{filename}.csv")
            df_qc.to_csv(qc_output_path, index=False)
            print(f"Saved QC metrics for {filename} as {qc_output_path}")
            
            # Generar y guardar las gráficas de QC
            qc_plot_path = os.path.join(qc_output_directory, f"qc_metrics_{filename}.png")
            plot_qc_metrics(df_qc, qc_plot_path)
            print(f"Saved QC plots for {filename} as {qc_plot_path}")
            
            # Filtrar los datos usando pegasus
            pg.filter_data(data)
            
            # Check if filtered data still contains integer counts
            if not np.issubdtype(data.X.dtype, np.integer):
                print(f"Warning: {filename} filtered data does not contain raw counts in .X")
            else:
                print(f"{filename} filtered data contains raw counts in .X")
            
            # Guardar los datos filtrados
            filtered_output_path = os.path.join(filtered_output_directory, f"filtered_{filename}")
            pg.write_output(data, filtered_output_path)
            print(f"Saved filtered data for {filename} as {filtered_output_path}")

# Definir los directorios de entrada y salida
input_directory = "/app/project/pipeline_colombia/1.5cell_group_filtered/Cell_groups"
qc_output_directory = "/app/project/pipeline_colombia/2.quality_metrics"
filtered_output_directory = "/app/project/pipeline_colombia/2.quality_control"
process_quality_control(input_directory, qc_output_directory, filtered_output_directory)
```
